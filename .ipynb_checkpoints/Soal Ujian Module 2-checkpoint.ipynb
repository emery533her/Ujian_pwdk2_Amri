### IMPUTE EMPTY ONES

```python
dictAgeWho = {}
for i in Titanic['who'].unique():
    meanagebywho = Titanic[Titanic['who']==i]['age'].mean()
    dictAgeWho[i] = meanagebywho
print(dictAgeWho)
```

```python
# def impute_age(col):
#     age = col['age']
    
#     if pd.isna(age):
#         return dictAgeWho[age]
#     else:
#         return age
```

```python
def impute_age(cols):
    age = cols['age']
    who = cols['who']
    
    if pd.isnull(age):
        return dictAgeWho[who]
    else:
        return age
```

```python
Titanic['age'] = Titanic.apply(impute_age, axis=1)
```

```python
for i in Titanic['who'].unique():
    filter3 = Titanic[Titanic['who']==i]['age']
    print(f"Banyak NaN {i} & mean: {sum(filter3.isnull())} & {filter3.mean()}")
```

### 9. Adding New Features/Columns/Variables

#### Buat DataFrame baru dengan column2 sebagai berikut
 - __id__ = id table ini (PK)
 - __mean_fare__ = Mean Fare
 - __std_fare__ = Std Fare
 - __batas_min_fare__ = Mean Fare - (3 x Std Fare)
 - __batas_max_fare__ = Mean Fare + (3 x Std Fare)
 
#### Dan data2nya ada 3 (fare setiap class yaitu first, second, and third class)

```python
Titanic['pclass'].unique()
```

```python
dictNum = {1:'First', 2:'Second', 3:'Third'}
```

```python
[dictNum[i] for i in Titanic['pclass'].unique()]
```

```python
mean_fare = []
std_fare = []
batas_min_fare = []
batas_max_fare = []
for num_class in Titanic['pclass'].unique():
    mean_class = Titanic[Titanic['pclass']==num_class]['fare'].mean()
    mean_fare.append(mean_class)
    std_class = Titanic[Titanic['pclass']==num_class]['fare'].std()
    std_fare.append(std_class)
    batas_min_fare.append(mean_class - 3* std_class)
    batas_max_fare.append(mean_class + 3* std_class)
TitanicOutCalc = pd.DataFrame(data = {'id':[1,2,3], 'mean_fare':mean_fare, 'std_fare':std_fare, 'batas_min_fare':batas_min_fare, 'batas_max_fare':batas_max_fare})
TitanicOutCalc
```

```python
translateDict = {1:3,2:1,3:2}
for i in [2,3,1]:
    min_thres = TitanicOutCalc[TitanicOutCalc['id']==i]['batas_min_fare'][i-1]
    max_thres = TitanicOutCalc[TitanicOutCalc['id']==i]['batas_max_fare'][i-1]
#     print(i, min_thres, max_thres, translateDict[i])
#     print(sum(~Titanic['fare'].between(min_thres, max_thres))/len(Titanic)*100)
    filtered_data = Titanic[Titanic['pclass']==translateDict[i]]['fare'].between(min_thres,max_thres)
    sum_out = sum(~filtered_data)
    len_out = len(filtered_data)
    print(f"Outlier % {dictNum[translateDict[i]]} : {(sum_out/len_out)*100}")
```

```python
sum_out = sum(~Titanic[Titanic['pclass']==3]['fare'].between(-21.658875,49.009975))
len_out = len(Titanic[Titanic['pclass']==3]['fare'].between(-21.658875,49.009975))
print(sum_out, len_out)
sum_out/len_out*100
```

#### Print outlier % of each class

#### Tambahkan column outliercalcId (FK) di dfTitanic

```python
def imputeOut(cols):
    pclass_id = cols['pclass']
    return {3:1,1:2,2:3}[pclass_id]
```

```python
Titanic['outliercalcId'] = Titanic.apply(imputeOut, axis=1)
Titanic.head()
```

#### Tambahkan column outlier (1 for yes and 0 for no)

```python
def imputeOutCat(cols):
    fare = cols['fare']
    i = cols['outliercalcId']
    min_thres = TitanicOutCalc[TitanicOutCalc['id']==i]['batas_min_fare'][i-1]
    max_thres = TitanicOutCalc[TitanicOutCalc['id']==i]['batas_max_fare'][i-1]
    
    if (fare > min_thres) & (fare < max_thres): 
#     fare.between(min_thres, max_thres):
        return 0
    else:
        return 1
```

```python
Titanic['outlier'] = Titanic.apply(imputeOutCat, axis=1)
Titanic.head()
```
